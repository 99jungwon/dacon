# 1. for 반복문

```python
# 리스트 numbers에서 홀수만 다른 리스트에 넣어서 반환하는 함수 return_odds

def return_odds(numbers):
    to_return = []
    for i in numbers:
        if i % 2 == 1:
            to_return.append(i)
    return to_return

return_odds([1, 2, 3, 4, 5, 6, 7])
```

# 2. while 반복문

# 3. break

```python
for i in range(10):
    if i == 5:
        print('5에 도착, 반복문 중단')
        break
    print('현재 값:', i)
```

# 4. continue

```python
# 자연수로 된 리스트 numbers가 주어졌을 때
# 7의 배수만 더해서 그 값을 반환하는 함수 sum_multiple_of_seven

def sum_multiple_of_seven(numbers):
    sum = 0
    for i in numbers:
        if i % 7 != 0:
            continue
        else:
            sum += i
    return sum

sum_multiple_of_seven([1, 14, 25, 47, 49, 63])
```

# 5. 반복문에서 인덱스의 활용(1)

```python
# 주어진 리스트 numbers가 있을 때 첫번째 짝수를 찾고 리스트에서
# 몇 번째 숫자인지 튜플로 반환하는 함수 first_even을 작성하세요.
# 만약 짝수가 없으면 0을 반환하세요.

def first_even(numbers):
    for i in range(len(numbers)):
        if numbers[i] % 2 == 0:
            return (numbers[i], i + 1)
    return 0

first_even([1, 3, 5, 8, 10, 21])
```

# 6. 반복문에서 인덱스의 활용(2)

### - 파이썬에서는 enumerate()을 사용하면 한번에 인덱스와 값을 모두 얻을 수 있습니다.

```python
# 주어진 리스트 numbers가 있을 때 첫번째 짝수를 찾고
# 리스트에서 몇 번째 숫자인지 튜플로 반환하는 함수 first_even을 작성하세요.
# 만약 짝수가 없으면 0을 반환하세요.

def first_even(numbers):
    for index, value in enumerate(numbers):
        if value % 2 == 0:
            return (value, index + 1)
    return 0

first_even([1, 3, 5, 8, 10, 21])
```

# 2. 리스트 컴프리헨션(2)

```python
# 리스트 numbers의 원소들의 제곱을 원소들로 하는 리스트 squares를 생성하세요.

numbers = [1, 2, 3, 4, 5]
squares = [i ** 2 for i in numbers]
print(squares)
```

# 3. 조건문을 활용한 리스트 컴프리헨션(1)


```python
# 0 부터 9까지의 숫자 중 짝수만 들어있는 리스트 even_numbers를 생성하세요.

even_numbers = [i for i in range(10) if i % 2 == 0]
print(even_numbers)
```

# 4. 조건문을 활용한 리스트 컴프리헨션 (2)

```python
# 자연수로 된 리스트 numbers가 주어졌을때 각 원소가 짝수이면 '짝수',
# 홀수이면 '홀수'로 바꾼 새로운 리스트를 반환하는 함수 odd_even을 작성하세요.

def odd_even(numbers):
    return ['짝수' if i % 2 == 0 else '홀수' for i in numbers]

odd_even([1, 2, 3, 4, 5])
```
