# loc함수는 판다스에서 라벨 기반의 인덱싱을 수행하기 위한 메소드이다.

```python
import pandas as pd

data = [20, 10, 40, 50, 60, 30, 70, 80]
index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']

series = pd.Series(data, index=index, name='my_series')

value_at_b = series.loc['b']
print(series)
print(value_at_b)
```

# iloc는 위치 기반 인덱싱을 수행하는 속성이다. iloc는 정수 기반의 인덱스를 사용하여 데이터에 접근한다.

```python
value_at_one = series.iloc[1]
value_at_one
```

# apply 메서드
 - apply() 함수는 판다스의 시리즈나 데이터프레임에 함수를 적용하는 메서드입니다.

```python
import pandas as pd

data = [1, 2, 3, 4, 5]
series = pd.Series(data)
series
```

```python
def custom_function(x):
    # 사용자 정의 함수의 내용을 작성
    if x >= 3:
        result = x + 2
    else:
        result = x + 4
    return result

result_series = series.apply(custom_function)
result_series
```

# apply 메서드와 lambda 함수
 - lambda함수는 익명 함수로, 한 줄로 표현되는 간단한 함수입니다.

```python
# 위에 코드와 똑같다
result_series = series.apply(lambda x: x + 2 if x >= 3 else x + 4)
result_series
```

# map 메서드
 - map 은 시리즈 내의 각 요소에 함술,ㄹ 적용하여 반환된 결과를 새로운 시리즈로 반환합니다.
 - apply는 데이터 프레임이나 시리즈 객체에 사용됩니다. 반면 map은 시리즈 객체에서만 사용됩니다.

```python
result_series = series.map(lambda x: x ** 2)
result_series
```

# replace 함수
 - replace함수는 정규표현식을 이용해 패턴에 일치하는 문자열을 찾아내고, 만약 패턴에 맞는 문자열을 찾아낸다면 미리 설정해둔 문자열로 데이터를 변경할 수 있다.

```python
data = ['apple', 'banana', 'cherry']
series = pd.Series(data)
series
```

```python
result_series = series.replace('^a', 'fruit', regex=True)
# 정규표현식을 사용할때는 regex = True를 사용해야 한다.
result_series
```

# cut함수
 - cut함수를 이용해서 0초과~30이하는 low, 30초과 60이하는 High로 값을 변경해 봅시다. 그리고 결과는 result_series 에 저장하세요.

```python
data = [10, 20, 30, 40, 50]
series = pd.Series(data)

# 구간을 정의합니다.
bins = [0, 30, 60]

# 라벨을 만들어줍니다.
labels = ['Low', 'High']

# cut 함수를 사용하여 값을 나눕니다.
result_series = pd.cut(series, bins, labels=labels)
result_series
```

# 시리즈에서 결측치 처리하기

```python
import pandas as pd

data = [20, None, 40, 50, None, 30, 70, None, 20, 50]
index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

series = pd.Series(data, index=index)
series
```

```python
# dropna()함수를 이용해 series의 결측치를 제거
result_series = series.dropna()
result_series
```

```python
# fillna()함수를 이용해서 결측치를 0으로 채우기
result_series = series.fillna(0)
result_series
```

# 중복되는 값 찾기
 - 첫 번째 나타난 중복값은 False로 제외하고 그 이후부터 중복된 값을 True로 표시한다.

```python
# duplicated()함수를 사용해 중복되는 값을 찾기
result_series = series.duplicated()
result_series
```

```python
# drop_duplicates() 함수를 이용해서 시리즈의 중복값을 제거
result_series = series.drop_duplicates()
result_series
```

# 데이터프레임 결합

```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

display(df1)
display(df2)
```

```python
# 데이터를 행방향으로 결합
concatenated = pd.concat([df1, df2], axis=0)
concatenated
```

```python
# 데이터를 열방향으로 결합
concatenated = pd.concat([df1, df2], axis=1)
concatenated
```

```python
# inner join 연산 merge()

# 첫 번째 데이터프레임 생성
df1 = pd.DataFrame({'ID': [1, 2, 3],
                    'name': ['철수', '영희', '민수']})

# 두 번째 데이터프레임 생성
df2 = pd.DataFrame({'ID': [1, 2, 4],
                    'age': [25, 30, 28]})

display(df1)
display(df2)

# '학생ID' 열을 기준으로 두 데이터프레임 조인
joined_df = pd.merge(df1, df2, on='ID', how='inner')
joined_df
```

```python
# left join(교집합 느낌)

# '학생ID' 열을 기준으로 Left 조인
left_joined_df = pd.merge(df1, df2, on='ID', how='left')
left_joined_df
```

```python
# outer join(합집합 느낌)

# '학생ID' 열을 기준으로 Outer 조인
outer_joined_df = pd.merge(df1, df2, on='ID', how='outer')
outer_joined_df
```

# 피벗 테이블

```python
# 예제 데이터
data = {'City': ['Seoul', 'Seoul', 'Seoul', 'Busan', 'Busan'],
        'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

print(df)
```

```python
# 피벗 테이블 생성
pivot_table = df.pivot_table(index='City', columns='Category', values='Value', aggfunc='sum')
pivot_table
```
